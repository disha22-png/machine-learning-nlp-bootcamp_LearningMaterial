### While Loop

```python
## While Loop - It continues to execute as long as the condition is true.

count = 0

while count<5:
    print(count)
    count=count+1
```

```python
count = 0
while count % 2 == 0:  # now as the 1 will be odd no. loop will stop at that point giving result as zero
    print(count)
    count = count + 1
```

### Loop Control Statement

```python
## break

for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
## continue = skips current iteration and continues with the next.

for i in range(10):
    if i%2 == 0:
        continue # it will skip that value which is even
    print(i)
```

```python
# Pass - is a null operations, does nothing

for i in range(5):
    if i== 3:
        pass
    print(i)
```

```python
## Nested Loop - Loop inside a loop

for i in range(3):
    for j in range(3):
        print(f"i:{i} and j:{j}")
```
